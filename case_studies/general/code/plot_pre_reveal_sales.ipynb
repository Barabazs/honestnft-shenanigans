```python
import pandas as pd
import matplotlib.pyplot as plt

COLLECTION_FOLDER = "svs"
COLLECTION_NAME = "SVS"

sales = pd.read_csv(
    "../../{}/data/{}_pre-reveal_sales.csv".format(COLLECTION_FOLDER, COLLECTION_FOLDER)
)

plt.scatter(x=sales["RANK"], y=sales["PRICE"], s=15, alpha=0.25)

plt.xlabel("Rarity Rank")
plt.ylabel("Price (Ether)")
plt.title("Pre-reveal Sales - {}".format(COLLECTION_NAME))
plt.savefig(
    "../../{}/figures/{}_price_vs_rank_pre-reveal.png".format(
        COLLECTION_FOLDER, COLLECTION_NAME
    ),
    dpi=300,
)
plt.show()
```
