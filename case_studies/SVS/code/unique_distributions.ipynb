```python
import pandas as pd
import matplotlib.pyplot as plt

TOP_N_TOKENS = 100

df_rarity = pd.read_csv("../data/distribution_raritytools.csv")
alt_df_rarity = pd.read_csv("../data/alt_distribution_raritytools.csv")

# Fix TOKEN_ID off by 1 error
df_rarity["TOKEN_ID"] = df_rarity["Unnamed: 0"] + 1
alt_df_rarity["TOKEN_ID"] = alt_df_rarity["Unnamed: 0"] + 1

ax = plt.subplot()
df_rarity = df_rarity[df_rarity["Rank"] <= TOP_N_TOKENS]
plt.scatter(x=df_rarity["TOKEN_ID"], y=df_rarity["Rank"], s=15, alpha=0.50)
plt.xlabel("Token ID")
plt.ylabel("Rarity Rank")
plt.title("Rarity Map - SVS (original)")
plt.savefig("../figures/rarity_map.png", dpi=300)
plt.show()


ax = plt.subplot()
alt_df_rarity = alt_df_rarity[alt_df_rarity["Rank"] <= TOP_N_TOKENS]
plt.scatter(
    x=alt_df_rarity["TOKEN_ID"], y=alt_df_rarity["Rank"], c="g", s=15, alpha=0.50
)
plt.xlabel("Token ID")
plt.ylabel("Rarity Rank")
plt.title("Rarity Map - SVS (altered)")
plt.savefig("../figures/altered_rarity_map.png", dpi=300)
plt.show()
```
