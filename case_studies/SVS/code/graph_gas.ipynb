```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.dates import DateFormatter
import matplotlib.ticker as mticker
import matplotlib.dates as mdates

BASE = "%Y-%m-%d %H:%M:%S"

data = pd.read_csv("../data/gas_data.csv")

fig, ax = plt.subplots()
dates = mdates.num2date(mdates.datestr2num(data["timestamps"]))

ax.plot(dates, data["base_fee"])
plt.title("Gas Price during SVS Mint")
plt.xlabel("UTC Time (Month, Day, Hour, Minute)")
plt.ylabel("Base Fee Per Gas (Gwei)")

ax.xaxis_date()
myFmt = DateFormatter("%m-%d %H:%M")
ax.xaxis.set_major_formatter(myFmt)

plt.xticks(rotation=45)

plt.autoscale()
plt.savefig("../figures/gas.png", dpi=300, bbox_inches="tight")
plt.show()
```
