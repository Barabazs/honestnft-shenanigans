```python
import requests

URL = "https://api.opensea.io/api/v1/bundles"
SSS_contract_address = "0xa7eE407497b2Aeb43580cabe2B04026b5419D1dc"
LIMIT = 50  # 50 is max
ETH_UNITS = 1e18


def GetEvents(offset):
    query_string = {
        "asset_contract_address": SSS_contract_address,
        "limit": LIMIT,
        "offset": offset,
    }

    response = requests.request("GET", URL, params=query_string)
    response_json = response.json()
    return response_json["bundles"]


bundles = []
offset = 0

while True:
    new_bundles = GetEvents(offset)
    bundles.extend(new_bundles)
    offset += LIMIT
    if len(new_bundles) < LIMIT:
        break

print("Length: ", len(bundles))
print("\n")


for i in range(0, len(bundles)):
    sell_orders = bundles[i]["sell_orders"]
    if len(sell_orders) != 0:
        print("Order Hash: ", sell_orders[0]["order_hash"])
        print("Created date: ", sell_orders[0]["created_date"])
        print("Price: ", float(sell_orders[0]["current_price"]) / ETH_UNITS)

    print("Username: ", bundles[i]["maker"]["user"]["username"])
    print("Seller Address: ", bundles[i]["maker"]["address"])

    assets = bundles[i]["assets"]

    for asset in assets:
        print("Token ID: ", asset["token_id"])

    print("\n")
```
