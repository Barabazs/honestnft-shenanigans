```python
import pandas as pd
import matplotlib.pyplot as plt

# batch 1 (https://etherscan.io/tx/0x61c4e32912df1dc831ddc57e73b1b01b2b0e4a2cd502377522ffaa3baa40c6a4)
# batch 2 (https://etherscan.io/tx/0x6d94a63896f1c9a73e8bf9edd4186576b3fa4f21c5c562e01671ca3f31588355)
# batch 3 (https://etherscan.io/tx/0xcbbdf802a4b5d5834705d94c880480304fadfa66f249b6623f2058c5c98f0991)


tx_ids = [
"0x61c4e32912df1dc831ddc57e73b1b01b2b0e4a2cd502377522ffaa3baa40c6a4",
"0x6d94a63896f1c9a73e8bf9edd4186576b3fa4f21c5c562e01671ca3f31588355",
"0xcbbdf802a4b5d5834705d94c880480304fadfa66f249b6623f2058c5c98f0991",    
]


mints = pd.read_csv("../data/superlativesecretsociety_minting.csv")
boolean_series = mints["txid"].isin(tx_ids)
mints = mints[boolean_series]

mints.to_csv("../data/giveaway_tokens_svs.csv")

plt.scatter(x=mints["TOKEN_ID"], y=mints["rank"], s=15, alpha=.25)
plt.xlabel('Token ID')
plt.ylabel('Rarity Rank')
plt.title("Giveaway Tokens - SVS")
plt.savefig("figures/giveaway.png", dpi=300)
plt.show()
```
